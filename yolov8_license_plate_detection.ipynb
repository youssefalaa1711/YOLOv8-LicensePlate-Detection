```python
def get_image_paths(folder_path): # Function that gets image paths
    image_extensions = ['.jpg', '.png', '.jpeg']
    image_paths = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if any(file.lower().endswith(ext) for ext in image_extensions):
                image_paths.append(os.path.join(root, file))
    return image_paths

model = YOLO('/content/milestone 3/best.pt') # Milestone 3 model that detects plates

result = model.predict('/content/3carpic.webp', save=True, show_conf=False, agnostic_nms=True,save_crop=True) # Input cars pic into model

cropped_image_paths = get_image_paths('/content/runs/detect/predict8/crops') # Using the function to get the paths of cropped images

model = YOLO('/content/milestone 4/best.pt') # Milestone 4 model that detects letters and numbers

for image_path in cropped_image_paths: # Loop on cropped images to detect letters in the plates
    result = model.predict(image_path, save=True, show_conf=False, agnostic_nms=True)
```
